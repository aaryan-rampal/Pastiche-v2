```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("starry-night.jpg", cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('triangle.jpg', cv2.IMREAD_GRAYSCALE)
img3 = cv2.imread('sharp-edge.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
def plot_image(img):
    if (len(img.shape) == 2):
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(img, interpolation='nearest')
    plt.axis('off')
    plt.show()
```

```python
plot_image(img3)
```
